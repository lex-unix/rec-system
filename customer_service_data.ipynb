# Filling the NaN values in rating column

This feature is essential in the proccess of training our recommender models, especially the rank model as it will try to predict the rating for the each operator retreived from retrieval model

```python
def calc_nan_values(df, column):
    return df[column].isna().sum()
```

```python
num_nan_rating =  calc_nan_values(df, 'customer_satisfaction_rating')
```

```python
num_nan_rating
```

## Ananlyze the size of dataset if decided to drop NaN values

We can see that our dataset is reduced significantly after dropping the rows that have NaN values in a ratings column

This is unacceptable for a dataset that is already small

```python
print(
    f'Reduced size of dataset by dropping rows that contatin NaN values in ratings col: {len(df) - num_nan_rating}'
)
```

```python
reduction_percentage = (len(df) - (len(df) - num_nan_rating)) / len(df) * 100

print(f'Reduction percentage: {reduction_percentage}')
```

## Using K-Nearest Neighbours to fill the NaN values in rating column

The features could be ticket priorty, ticket status, ticket type and producet purchased

```python
df_knn = df[['ticket_priority', 'ticket_status', 'ticket_type', 
             'product_purchased', 'ticket_subject', 'customer_satisfaction_rating']].copy()
```

```python
le = LabelEncoder()

categorical_cols = ['ticket_status', 'ticket_priority', 'ticket_subject', 'ticket_type', 'product_purchased']
for col in categorical_cols:
    df_knn[col] = le.fit_transform(df_knn[col])
```

```python
imputer = KNNImputer(n_neighbors=5)

df_knn_filled = imputer.fit_transform(df_knn)

df_knn_filled = pd.DataFrame(df_knn_filled, columns=df_knn.columns)
```

```python
df['customer_satisfaction_rating'] = df_knn_filled['customer_satisfaction_rating']
```

```python
df['customer_satisfaction_rating'] = (df['customer_satisfaction_rating'] * 2).round() / 2
```

```python
df
```

## Generating operator names

```python
operator_names = set()
total_operators = 1000
fake = Faker()

while len(operator_names) != total_operators:
    operator_names.add(fake.name())
```

```python
operator_names = list(operator_names)
```

```python
df['operator'] = np.random.choice(operator_names, size=len(df))
```
